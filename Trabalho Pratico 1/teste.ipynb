<p> 1- Geração de um checkerboard utilizando a biblioteca NumPy (sem utilizar
estruturas de repetição na linguagem).,</p>
<p>⇨ Obs.: Utilize a biblioteca MatplotLib para exibir os resultados de cada etapa.</p>
<p>a. Primeiramente, construa um padrão preto (0’s) e outro branco (1’s) de 10x10 cada e armazene em variáveis B (0’s) e W (1’s).<br>
 Utilize as funções ‘ones’ e ‘zeros’ da NumPy.</p>

```python
import numpy as np
import matplotlib.pyplot as plt

B = np.zeros((10, 10))

W = np.ones((10, 10))

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(B, cmap='gray', vmin=0, vmax=1)
plt.title('Padrão Preto (0\'s)')


plt.subplot(1, 2, 2)
plt.imshow(W, cmap='gray', vmin=0, vmax=1)
plt.title('Padrão Branco (1\'s)')


plt.show()
```

<p>b. Utilize os padrões usados no item a para construir um padrão como o mostrado
abaixo.</p>
<p>Para isto, utilize uma das funções de concatenação da NumPy (concatenation,hstack ou c_).</p>

```python

BW = np.concatenate((B, W), axis=1)

plt.imshow(BW, cmap='gray', vmin=0, vmax=1)
plt.title('Padrão Black and White (BW)')
plt.show()
```

<p>c. De posse do padrão BW, gere um padrão de repetição horizontal composto por 5x o padrão BW.<br>
Para isto, utilize a função tile da NumPy e salve o resultado em ums variável BW_tiled, por exemplo.</p>

```python
BW_tiled = np.tile(BW, (1, 5))

plt.imshow(BW_tiled, cmap='gray', vmin=0, vmax=1)
plt.title('Padrão de Repetição Horizontal')
plt.show()
```

<p>d. Gere os procedimentos descritos nos itens b) e c) para um padrão invertido WB (branco e preto) gerando uma variável<br>
 de nome WB_tiled, por exemplo.</p>

```python
WB = np.concatenate((W, B), axis=1)

plt.imshow(WB, cmap='gray', vmin=0, vmax=1)
plt.title('Padrão Invertido (WB)')
plt.show()

WB_tiled = np.tile(WB, (1, 5))

plt.imshow(WB_tiled, cmap='gray', vmin=0, vmax=1)
plt.title('Padrão de Repetição Horizontal Invertido (WB_tiled)')
plt.show()
```

<p>e. Concatene verticalmente as variáveis BW_tiled e WB_tiled empilhando 5 cópias desse novo padrão<br>
 (utilize os comandos hstack e tile para realizar tais procedimentos) gerando um padrão de tabuleiro<br>
 de 100x100 de zeros e uns.</p>

```python

intercalar_horizontal = np.vstack((BW_tiled, WB_tiled))

pattern_final = np.tile(intercalar_horizontal, (5, 1))

plt.imshow(pattern_final, cmap='gray', vmin=0, vmax=1)
plt.title('Padrão de Tabuleiro de 100x100')
plt.show()
```

<p>2- Gere e exiba uma versão em cinza médio (127) para todos pixels brancos da sua
imagem gerada no item 1.e.<br>
→ Obs.: Converta o array para o tipo uint8 e use o colormap &#39;gray&#39; para exibição junto a função imshow da MatplotLib,<br>
 você também terá que modificar os parâmetros vmin e vmax desta função para exibir a imagem corretamente (caso contrário, você<br>
continuará visualizando a mesma imagem gerada antes de modificar os pixels brancos de 255 para 127).</p>

```python
pattern_final[pattern_final == 1] = 127


pattern_final = pattern_final.astype(np.uint8)


plt.imshow(pattern_final, cmap='gray', vmin=0, vmax=255)
plt.title('Padrão de Tabuleiro de 100x100 - Cinza Médio')
plt.show()
```

<p>3-De posse do tabuleiro gerado no item 1. , gere versões em 5 cores diferentes:vermelho, verde, azul, amarelo e magenta.</p>
<p>⇨ OBS.:o Não utilize laços de repetição</p>
<p>o Converta todas os arrays para o tipo uint8, ou seja, com valor mínimo de 0 e
máximo de 255.</p>
<p>o Utilize a função de subplot para exibir todos os 5 padrões coloridos gerados na <br>
 mesma figura incluindo a versão original, totalizando 6 imagens (2 linhas com 3 imagens por linha).</p>
<p>o Para cada subplot adicione um título referente ao que está sendo exibido.</p>

```python

red_pattern = np.stack((np.where(pattern_final == 127, 255, 0), np.zeros_like(pattern_final), np.zeros_like(pattern_final)), axis=-1)
green_pattern = np.stack((np.zeros_like(pattern_final), np.where(pattern_final == 127, 255, 0), np.zeros_like(pattern_final)), axis=-1)
blue_pattern = np.stack((np.zeros_like(pattern_final), np.zeros_like(pattern_final), np.where(pattern_final == 127, 255, 0)), axis=-1)
yellow_pattern = np.stack((np.where(pattern_final == 127, 255, 0), np.where(pattern_final == 127, 255, 0), np.zeros_like(pattern_final)), axis=-1)
magenta_pattern = np.stack((np.where(pattern_final == 127, 255, 0), np.zeros_like(pattern_final), np.where(pattern_final == 127, 255, 0)), axis=-1)

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(pattern_final, cmap='gray', vmin=0, vmax=255)
plt.title('Original')


plt.subplot(2, 3, 2)
plt.imshow(red_pattern)
plt.title('Vermelho')


plt.subplot(2, 3, 3)
plt.imshow(green_pattern)
plt.title('Verde')


plt.subplot(2, 3, 4)
plt.imshow(blue_pattern)
plt.title('Azul')


plt.subplot(2, 3, 5)
plt.imshow(yellow_pattern)
plt.title('Amarelo')


plt.subplot(2, 3, 6)
plt.imshow(magenta_pattern)
plt.title('Magenta')


plt.show()

```

<p>1-Utilize a OpenCv para ler a imagem do logo da Coca-Cola disponibilizada junto

```python
import cv2

image = cv2.imread('coca-cola-logo.png')

if image is None:
    print("Não foi possível carregar a imagem.")
else:
    print("Imagem carregada com sucesso.")

```

<p>2. (0,25 pts) Exiba a imagem lida com a função do MatplotLib imshow no espaço de cor
RGB.</p>

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.imshow(image_rgb)
plt.title('Coca-Cola Logo')
plt.axis('off')
plt.show()
```

<p>3-Utilize a biblioteca NumPy para gerar uma versão da imagem lida com as cores
trocadas, por exemplo, troque o que está branco para vermelho<br>
e o que está vermelho para verde</p>

```python

white_lower = np.array([200, 200, 200], dtype=np.uint8)
white_upper = np.array([255, 255, 255], dtype=np.uint8)
red_lower = np.array([0, 0, 150], dtype=np.uint8)
red_upper = np.array([100, 100, 255], dtype=np.uint8)

white_mask = cv2.inRange(image_rgb, white_lower, white_upper)
red_mask = cv2.inRange(image_rgb, red_lower, red_upper)

text_mask = cv2.bitwise_or(white_mask, red_mask)

blank_image = np.zeros_like(image_rgb)

blank_image[text_mask > 0] = [255, 0, 0]

inverse_text_mask = cv2.bitwise_not(text_mask)

blank_image[inverse_text_mask > 0] = [0, 255, 0]

plt.imshow(blank_image)
plt.title('Coca-Cola Logo - Modificado')
plt.axis('off')
plt.show()
```
